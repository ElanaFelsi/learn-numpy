# No Matter how you slice it

```python
%pylab inline
import re

def parameters(line):
    i, x, y, m, n = [int(x) for x in re.findall('\d+', line)]
    return i, x, y, m, n

def solve(s):
    fabric = np.zeros((1010, 1010), dtype=float)
    file = s.split("\n")
    for line in file:
        i, x, y, m, n= parameters(line)
        fabric[x:x+m, y:y+n] += 1
    
    imshow(fabric)
    return np.count_nonzero(fabric>1)

assert solve(open("day3.txt").read()) == 104712

print("OK")
```

```python
%pylab inline
import re

def parameters(line):
    i, x, y, m, n = [int(x) for x in re.findall('\d+', line)]
    return i, x, y, m, n

def solve2(s):
    fabric = np.zeros((1010, 1010), dtype=float)
    file = s.split("\n")
    for line in file:
        i, x, y, m, n= parameters(line)
        fabric[x:x+m, y:y+n] += 1

    for line in file:
        i, x, y, m, n= parameters(line)
        if (fabric[x:x+m, y:y+n] == 1).all():
            return i


assert solve2(open("day3.txt").read()) == 840

print("OK")
```
