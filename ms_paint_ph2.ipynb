```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import imageio
```

```python
pallete_img = imageio.imread("palette.png")
plt.axis('off')
pallete = pallete_img[0,:, :3]
print(pallete.shape, pallete.dtype)
plt.imshow(pallete_img)
```

```python
target = imageio.imread("starry-night-800.jpg")
plt.imshow(target)
print(target.shape, target.dtype)
```

```python
copy = np.full_like(target, 255)
```

```python
A = 100
for x in range(1000000):
    i, j = np.random.randint(634), np.random.randint(800)
    rgb = pallete[np.random.randint(len(pallete))]
    target_rgb = target[i:i+4, j:j+4]
    avg = np.average(target_rgb, axis=(0, 1))
    if np.all(abs(avg - rgb) <= A):
        copy[i:i+4, j:j+4] = rgb
plt.imshow(copy)
```

```python
imageio.imsave(f"copy.png", copy)
```
