# Some Basic Plotting

```python
plot(x,y)
```

```python
help(plot)
```

```python
plot(x,y,'bp-')
```

```python
plot(x,y,'bp-', x, cos(x), 'go-')
```

```python
plot(x,y)
plot(x,y*2)
```

```python
n = 5
for i in range(n):
    plot(x,y*i)
```

```python
n = 5
for i in range(n):
    plot(x,y*i, label=f"y * {i}")
legend()
```

```python
for i in range(n):
    subplot(n, 1, i+1)
    plot(x,y*i)
    title(f"y * {i}")
```
