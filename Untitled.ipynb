```python
import numpy as np
from matplotlib import pyplot as plt
import imageio
```

```python
img = imageio.imread("seam_carving_img.jpg")/255
plt.imshow(img)
```

```python
def calc_energy(img):
    x=abs(np.gradient(img,axis=0))
    y=abs(np.gradient(img,axis=1))
    i=np.array(x+y)
    return np.mean(i,-1)
energy_map = calc_energy(img)
plt.imshow(energy_map)
```

```python
def min_roll(row, c):
        row = np.append(row, [1,1]) 
        roll_mat = np.array([row, np.roll(row, 1), np.roll(row, 2)])
        print(roll_mat)
        #roll_mat = roll_mat[:, 1:c+1]
        min_row = np.min(roll_mat, axis=0)
        idx = np.arange(c+2) - (np.argmin(roll_mat, axis=0))
        return min_row[1:c+1], idx[1:c+1]
        
def calc_min_mat(energy_map):
    r, c = energy_map.shape 
    min_energy_map = np.zeros(energy_map.shape)
    backtrack = np.zeros(energy_map.shape)
    min_energy_map[0], backtrack[0] = min_roll(energy_map[0], c)
    
    for i in range(1,1):
        min_energy_map[i], backtrack[i] = min_roll(min_energy_map[i-1] + energy_map[i], c)
    
    plt.imshow(min_energy_map)
    plt.figure()
    print(backtrack)
    plt.imshow(backtrack)    
        
calc_min_mat(energy_map)        
```
